## 1. Chargement & aperçu

```python
import polars as pl
pl.Config.set_tbl_rows(20)
pl.Config.set_fmt_str_lengths(60)
pl.Config.set_tbl_width_chars(200)
```

```python
read_csv_options = {"infer_schema_length": None}
df_container = pl.read_csv("./tables athena/dict_dico_container.csv", **read_csv_options)
df_field     = pl.read_csv("./tables athena/difi_dico_field.csv",     **read_csv_options)
df_structure = pl.read_csv("./tables athena/dist_dico_structure.csv", **read_csv_options)
df_entt      = pl.read_csv("./tables athena/lien_link_entt.csv",      **read_csv_options)
df_source    = pl.read_csv("./tables athena/diso_dico_source.csv",    **read_csv_options)
df_usage     = pl.read_csv("./tables athena/usag_usage.csv",     **read_csv_options)


tables = {
    "source":    df_source,
    "container": df_container,
    "structure": df_structure,
    "field":     df_field,
    "link":      df_entt,
        "usage":     df_usage,

}
```

## 2. Volumétrie & structure de chaque table

```python
print(f"{'Table':<15} {'Lignes':>8} {'Colonnes':>10} {'Nulls total':>14} {'Str cols':>10} {'Num cols':>10}")
for name, df in tables.items():
    n_rows    = df.shape[0]
    n_cols    = df.shape[1]
    n_nulls   = sum(df[c].null_count() for c in df.columns)
    n_str     = sum(1 for c in df.columns if df[c].dtype == pl.Utf8)
    n_num     = sum(1 for c in df.columns if df[c].dtype in (pl.Int64, pl.Float64, pl.Int32))
    print(f"{name:<15} {n_rows:>8} {n_cols:>10} {n_nulls:>14} {n_str:>10} {n_num:>10}")
```

```python
def schema_detail(df, name):
    print(f"  {name.upper()}")
    print(f"  {'Colonne':<38} {'Type':>12} {'Nulls':>8} {'%Null':>8} {'Uniques':>10}")
    for col in df.columns:
        dtype   = str(df[col].dtype)
        n_null  = df[col].null_count()
        pct     = round(n_null / df.shape[0] * 100, 1)
        n_uniq  = df[col].n_unique()
        print(f"  {col:<38} {dtype:>12} {n_null:>8} {pct:>7}% {n_uniq:>10}")

for name, df in tables.items():
    schema_detail(df, name)
```
